## importing the Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm #svm --> support vector machine
from sklearn.metrics import accuracy_score
```

## Data Collection and Analysis
### PIMA diabetes.csv

```python
###### Loading the diabetes.csv Dataset to a pandas DataFrame ##########

diabetes_dataset = pd.read_csv("diabetes.csv")
```

```python
diabetes_dataset.head() #outcome 1 mean ----->Diabetic,,,,,,the all are females
```

#### 0 ----> non-diabetic
#### 1 ----> diabetic

```python
diabetes_dataset.groupby("Outcome").mean()
```

  ## separating  the data and labels

```python
X = diabetes_dataset.drop(columns = "Outcome" ,axis=1)
Y = diabetes_dataset["Outcome"]
```

```python
X.shape
```

```python
Y.shape
```

```python
X.head(3)
```

```python
Y.head(3)
```

## Data standardization    using  (from sklearn.preprocessing import StandardScaler) StandardScaler function

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
standardized_data
```

```python
X = standardized_data
#Y = diabetes_dataset["Outcome"]
```

```python
print(X)
print(Y)
```

### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
print(diabetes_dataset.shape) # actual data set
```

### Traning the model

```python
classifier = svm.SVC(kernel="linear")
```

## Traning the Support Vector Machine classifier

```python
classifier.fit(X_train, Y_train)
```

## model evaluation

### Accuracy Score

```python
# Accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("the accuracy score of the traning data : ",traning_data_accuracy)
```

```python
# Accuracy score on the testing data
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("the accuracy score of the testing data : ",testing_data_accuracy)
```

# making a predictionn system `

```python
input_data = (9,156,86,28,155,34.3,1.189,42)

# chandging the input data into numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we prediciding from one instance

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)

print(prediction)

if(prediction[0] == 0):
    print("the persion is not diabetic :",prediction)
else:
        print("the persion is diabetic :",prediction)
```
